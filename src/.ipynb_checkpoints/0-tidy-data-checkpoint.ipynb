```python
import pandas as pd
```

```python
df = pd.read_csv('../data/ellipsis_raw_rater_scores_anon_all_essay.csv')
```

```python
df_1 = df[['Filename', 'Text', 'Rater_1', 'Overall_1', 'Cohesion_1', 'Syntax_1',
           'Vocabulary_1', 'Phraseology_1', 'Grammar_1', 'Conventions_1',
           'Identifying_Info_1']]
df_2 = df[['Filename', 'Text','Rater_2', 'Overall_2', 'Cohesion_2', 'Syntax_2',
           'Vocabulary_2', 'Phraseology_2', 'Grammar_2', 'Conventions_2',
           'Identifying_Info_2']]

columns = ['Filename', 'Text', 'Rater', 'Overall', 'Cohesion', 'Syntax',
           'Vocabulary', 'Phraseology', 'Grammar', 'Conventions',
           'Identifying_Info']

df_1.columns = columns
df_2.columns = columns

df = pd.concat([df_1, df_2])
```

```python
df = pd.melt(df, id_vars=['Filename', 'Text', 'Rater'], 
        value_vars=['Overall', 'Cohesion', 'Syntax','Vocabulary', 'Phraseology', 'Grammar', 'Conventions', 'Identifying_Info'],
        value_name='criterion', col_level=None, ignore_index=True)
```

```python
df.to_csv('../data/ellipse_raw_rater_scores_tidy.csv')
```
